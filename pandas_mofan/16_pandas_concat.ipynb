```python
import pandas as pd 
import numpy as np
```

# contenating

```python
df1 = pd.DataFrame(np.ones((3,4))*0 , columns=['a','b','c','d'])
df2 = pd.DataFrame(np.ones((3,4))*1 , columns=['a','b','c','d'])
df3 = pd.DataFrame(np.ones((3,4))*2 , columns=['a','b','c','d'])
print(df1)
print(df2)
print(df3)
```

## 上下合并

```python
res = pd.concat([df1,df2,df3],axis=0)
res
```

```python
res = pd.concat([df1,df2,df3],axis=0,ignore_index=True)
res
```

```python
## join, ['inner' ,'outer']
```

```python
df1 = pd.DataFrame(np.ones((3,4))*0 , columns=['a','b','c','d'],index=[1,2,3])
df2 = pd.DataFrame(np.ones((3,4))*1 , columns=['b','c','d','e'],index=[2,3,4])

res = pd.concat([df1,df2])
print(res)


res1 = pd.concat([df1,df2],join='inner')
print(res1)

res2 = pd.concat([df1,df2],join='inner',ignore_index=True)
print(res2)
```

```python
res1 = pd.concat([df1,df2],join='outer')
print(res1)

res2 = pd.concat([df1,df2],join='outer',ignore_index=True)
print(res2)
```

# append

```python
df1 = pd.DataFrame(np.ones((3,4))*0 , columns=['a','b','c','d'])
df2 = pd.DataFrame(np.ones((3,4))*1 , columns=['a','b','c','d'])
df3 = pd.DataFrame(np.ones((3,4))*2 , columns=['a','b','c','d'])

res=df1.append(df2,ignore_index=True)
res
```

```python
res=df1.append([df2,df3],ignore_index=True)
res
```

```python
df1 = pd.DataFrame(np.ones((3,4))*0 , columns=['a','b','c','d'])
s = pd.Series([1,2,3,4],index=['a','b','c','d'])

res = df1.append(s,ignore_index=True)
res
```
